**4. Config the axes**

Use the above generated dataset, let set the name for x-axis, y-axis and the legend of your chart.
Suggested reading: [link](https://www.tutorialspoint.com/matplotlib/matplotlib_axes_class.htm)

```
x = np.arange(0,100)
y = np.random.randint( low = 100 , high = 1000 , size = 100)

fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x,y,'ys-')

ax.legend(labels = ( 'smartphone'))
plt.title('graph line')
plt.xlabel('X axis ')
plt.ylabel('Y axis ')
plt.show()
```

Let draw the above two lines y1 and y2. The chart must has:


1.   names of x-axis and y-axis
2.   chart legend
3.   different colors for the lines of y1 and y2
4.   different markers for the lines of y1 and y2





```
x = np.arange(0,100)
y1 = np.random.randint( low = 0 , high = 1000 ,size = 100 )
y2 = np.random.randint( low = 500, high = 1000 , size = 100)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x,y1,'bo:')
l2 = ax.plot(x,y2,'cD:')
ax.legend(labels = ('pc', 'console'), loc = 'lower center')
ax.set_title("Advertisement effect on sales")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```
